```julia
using Plots
default(xlabel="Time, sec");
```

```julia
L0 = CsvToStruct("Link0");
L1 = CsvToStruct("Link1");
```

```julia
L0.Names
```

```julia
plot(ylims=(0,6),title="Uplink Occultation")
plot!(L0.Time,3*L0.PathIsOcculted,label="Path Occulted")
plot!(L0.Time,2*L0.TxOcculted,label="Occulted at Tx")
plot!(L0.Time,1*L0.RxOcculted,label="Occulted at Rx")
```

```julia
plot(ylims=(0,6),title="Downlink Occultation")
plot!(L1.Time,3*L1.PathIsOcculted,label="Path Occulted")
plot!(L1.Time,2*L1.TxOcculted,label="Occulted at Tx")
plot!(L1.Time,1*L1.RxOcculted,label="Occulted at Rx")
```

```julia
plot(ylims=(0,3),title="Total Link Occultation")
plot!(L0.Time,1*L0.PathIsOcculted,label="Uplink")
plot!(L1.Time,2*L1.PathIsOcculted,label="Downlink")
```

```julia
plot(title="Ground Antenna Gains")
plot!(ylabel="dB")
plot!(L0.Time,L0.TxAntGain,label="GSFC Uplink")
plot!(L1.Time,L1.RxAntGain,label="WS1 Downlink")
```

```julia
plot(title="S/C Antenna Gains")
plot!(ylabel="dB")
plot!(L0.Time,L0.RxAntGain,label="S-Band Omni")
plot!(L1.Time,L1.TxAntGain,label="X-Band HGA")
```

```julia
plot(title="Delay Times")
plot!(ylabel="sec")
plot!(L0.Time,L0.Delay,label="Uplink")
plot!(L1.Time,L1.Delay,label="Downlink")
```

```julia
plot(title="Carrier-to-Noise Ratio")
plot!(ylabel="dB")
plot!(L0.Time,L0.CNR,label="Uplink")
plot!(L1.Time,L1.CNR,label="Downlink")
hline!([0],color=:gray)
```

```julia
plot(title="Range")
plot!(ylabel="km")
plot!(L0.Time,L0.Range/1000,label="Uplink")
plot!(L1.Time,L1.Range/1000,label="Downlink")
```

```julia
plot(title="Range Rate")
plot!(ylabel="m/sec")
plot!(L0.Time,L0.RangeRate,label="Uplink")
plot!(L1.Time,L1.RangeRate,label="Downlink")
```

```julia
plot(title="Doppler Shift")
plot!(ylabel="kHz")
plot!(L0.Time,L0.Doppler/1000,label="Uplink")
plot!(L1.Time,L1.Doppler/1000,label="Downlink")
```

```julia
plot(title="Effective Isotropic Radiated Power")
plot!(ylabel="dBW")
plot!(L0.Time,L0.EIRP,label="Uplink")
plot!(L1.Time,L1.EIRP,label="Downlink")
```

```julia
plot(title="Free Space Path Loss")
plot!(ylabel="dB")
plot!(L0.Time,L0.FSPL,label="Uplink")
plot!(L1.Time,L1.FSPL,label="Downlink")
```
